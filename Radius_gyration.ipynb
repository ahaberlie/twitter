```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image
from glob import glob
from pingel_twitter import load_twitterlog
from mpl_toolkits.basemap import Basemap
from math import sqrt
```

```python
#calculate the centroid of the daily tweet locations
def calculate_weighted_centroid(x,y):
    x_cent = np.mean(x)
    y_cent = np.mean(y)
    return x_cent,y_cent
    
#calculate radius of gyration http://www.nature.com/nature/journal/v453/n7196/full/nature06958.html
def radius_gyration(x,y):
    x_cent,y_cent = calculate_weighted_centroid(x,y)
    distance_sum = 0.0
    for x1,y1 in zip(x,y):
        distance = sqrt((x_cent-x1)**2 + (y_cent-y1)**2)
        distance_sum += distance
    MSE = (1.0/len(x)) * (distance_sum**2)
    return x_cent,y_cent,sqrt(MSE)

Image(filename="C:\\users\\ahaberlie\\desktop\\rad_gyration.png")
```

```python
#get list of daily twitter data
d = glob("E:\\Twitter\\*.txt")

#read in first day
dataset = load_twitterlog(d[0])

#use first day's user ids to build a new dataframe
build_dataframe = pd.DataFrame(np.unique(dataset['user_id']),columns=['user_id'])

#set the user_id as the label.. this can speed up searching for user id
build_dataframe = build_dataframe.set_index('user_id')

#instantiate a basemap object.. need this to project lat/lon data to run distance calculations
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
    projection='lcc',lat_1=33,lat_2=45,lon_0=-95,resolution='i')

build_dataframe[0:10]
```

```python
#loop through all the daily twitter data
for file_name in d:
    print "loading file....", file_name
    dataset = load_twitterlog(file_name)
    
    #get count of unique user ids
    length = len(np.unique(dataset['user_id']))
    
    #get the date from the filename
    date_label = file_name[11:-4]
    
    #build array to hold daily statistics of gyration radius
    temp_array = []
    
    #creates an object where all the data for each user is grouped into a list attribute
    grouped = dataset.groupby('user_id')
    
    #used to monitor the processing speed
    curr_count = 0
    
    #loop through each user id group
    #group[0] is the user_id
    #group[1] is all the data for that user on that day
    for group in grouped:
        
        #get the lat lon list for each of the user's tweets
        x,y = np.array(group[1]['lon']), np.array(group[1]['lat'])
        
        #make sure they have at least 2 tweets or the distance calculation is useless
        if len(x) >= 2:
            
            #project
            x,y = m(x,y)
            
            #calculate radius of gyration, also return weighted centroid
            cx,cy,rog = radius_gyration(x,y)
            
            #build list for user including their average location and gyration radius that day
            temp_array.append((group[0],cx,cy,rog))
            
            #used to monitor the processing speed
            print date_label, "on user",curr_count,"out of",len(grouped)
        
        #user to monitor the processing speed
        curr_count += 1
    
    #build a dataframe from the python list
    #I perform an inline list to array conversion
    temp_dataframe = pd.DataFrame(np.array(temp_array),columns=['user_id',date_label+"_x",date_label+"_y",date_label+"_rog_value"])
    
    #set the index to the user id to speed up searching
    temp_dataframe = temp_dataframe.set_index('user_id')
    
    #perform a left join with the master dataframe and the temporary (daily) dataframe
    build_dataframe = build_dataframe.join(temp_dataframe,how='left')
    
    #delete the dataset for this iteration
    dataset = None
```
