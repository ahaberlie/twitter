First part of the implementation described in this paper: http://www.nature.com/nature/journal/v453/n7196/full/nature06958.html
        
This rotates a group of points so that the major axis (moment of inertia) is aligned with the x-axis

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import time
import sys
import time

from IPython.display import display, clear_output
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

```python
def calculate_weighted_centroid(x,y):
    x_cent = np.mean(x)
    y_cent = np.mean(y)
    return x_cent,y_cent
    
def RMSE(pts):
    pts_cent = np.zeros((len(pts)))
    pts_cent[:] = np.mean(pts)
    return math.sqrt(mean_squared_error(pts,pts_cent))

##calculate angle of major axis relative to x axis
def moment_inertia(x,y):    
    Ixx = np.sum(y**2)    
    Iyy = np.sum(x**2)    
    Ixy = Iyx = -np.sum(x*y)    
    
    mu = np.sqrt(4*Ixy*Iyx + Ixx**2 - 2*Ixx*Iyy + Iyy**2)    
    
    first_term = -Ixy*(.5*Ixx - .5*Iyy + .5*mu)**-1    
    second_term = np.sqrt(1 / ( 1 + (Ixy**2 / (.5*Ixx - .5*Iyy + .5*mu)**2)))    
    
    theta = np.arccos(first_term * second_term) #theta in radians
    return theta

##rotate vector
def rotate_vector(x, y, angle, xc, yc):       
   x = x - xc    
   y = y - yc      
   
   cos_theta = math.cos(angle)    
   sin_theta = math.sin(angle)    
   
   nx = x*cos_theta - y*sin_theta    
   ny = x*sin_theta + y*cos_theta    
   
   nx = nx + xc    
   ny = ny + yc    
   
   return [nx, ny]
```

```python
##create randomly placed test points
##this mimics the users locations throughout the day
x = []
y = []
for i in range(30):
    random.seed(time.time()*10+i+i)
    x.append(random.random()*10+i)
    random.seed(time.time()+i)
    y.append(random.random()*10+i)

x = np.array(x)
y = np.array(y)

cx,cy = calculate_weighted_centroid(x,y)

plt.scatter(x,y)
plt.plot(cx,cy,'*r',markersize = 10)
plt.xlim(0,50)
plt.ylim(0,50)
```

```python
theta = moment_inertia(x,y)
print math.degrees(theta)
```

```python
nx,ny = rotate_vector(x,y,-theta,cx,cy)

plt.scatter(nx,ny)
plt.plot(cx,cy,'*r',markersize = 10)
plt.xlim(0,50)
plt.ylim(0,50)
```

What if we wanted to align with the y-axis?

```python
nx,ny = rotate_vector(nx,ny,-math.radians(90),cx,cy)

plt.scatter(nx,ny)
plt.plot(cx,cy,'*r',markersize = 10)
plt.xlim(0,50)
plt.ylim(0,50)
```

```python
f, ax = plt.subplots()
for i in range(180):
    time.sleep(.001)
    nx,ny = rotate_vector(nx,ny,-math.radians(1),cx,cy)
    ax.scatter(nx,ny)
    ax.plot(cx,cy,'*r',markersize = 10)
    plt.xlim(0,50)
    plt.ylim(0,50)
    display(f)
    clear_output(wait=True)
```
